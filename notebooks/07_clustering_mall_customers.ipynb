```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Age": [19, 21, 20, 23, 31, 22, 35, 45, 29, 30],
    "Annual Income (k$)": [15, 16, 17, 20, 22, 35, 60, 75, 45, 50],
    "Spending Score": [39, 81, 6, 77, 40, 76, 6, 94, 50, 60]
}

df = pd.DataFrame(data)

X = df[["Annual Income (k$)", "Spending Score"]]
model = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = model.fit_predict(X)

plt.scatter(df["Annual Income (k$)"], df["Spending Score"], c=df["Cluster"])
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.show()
```
